```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

%matplotlib inline 
```

```python
large = pd.read_csv("DF17/data.csv", sep="\t")
large.dropna(inplace = True)
```

```python
dest = pd.read_csv("DF17/dest.csv")
dest_only = dest[['srch_destination_id','srch_destination_name']]
dest_only
```

```python
large = large[["date_time", "hotel_country", "is_booking", "srch_destination_id", 'user_location_city']]
large = large[large.is_booking != 0]
large['date_time'] = pd.to_datetime(large['date_time'])
large['date_time'] = large['date_time'].dt.date
large['Month'] = pd.DatetimeIndex(large.date_time).month
large
```

```python
USMerge = large.merge(dest_only,how='left',on='srch_destination_id')
#USMerge = USMerge[['Month','srch_destination_name']
USMerge = USMerge[USMerge.hotel_country == "UNITED STATES OF AMERICA"]
USMerge
```

```python
b = USMerge.groupby('srch_destination_name').size()
TopUS = b.order(ascending=False)
TopUS
```

```python
country_names = large.hotel_country.unique()
country_names
```

```python
Month_Country_df = large[['hotel_country','Month']]
Month_Country_df.sort("Month")
```

```python
p = Month_Country_df.groupby('hotel_country').size()
Top5_Country_Searches = p.order(ascending=False)[0:5]
Top5_Country_Searches
```

```python
Top10_Country_Searches = p.order(ascending=False)[0:10]
Top10_Country_Searches
Top10_Country_Searches.sum()
```

```python
All = Month_Country_df.groupby(['hotel_country','Month']).size()#.agg(['count'])

Test = pd.DataFrame({'count' : Month_Country_df.groupby(['hotel_country','Month']).size()}).reset_index()#.agg(['count'])

USA = Test[Test.hotel_country == "UNITED STATES OF AMERICA"]
CAN = Test[Test.hotel_country == "CANADA"]
MEX = Test[Test.hotel_country == "MEXICO"]
ITA = Test[Test.hotel_country == "ITALY"]
UK = Test[Test.hotel_country == "UNITED KINGDOM"]
#Month_Country_df
```

```python
for country_name in country_names:
#     print(country_name)
#     print(All[country_name])
    if country_name != "UNITED STATES OF AMERICA" and country_name != "CANADA" and country_name != "MEXICO" and country_name != "ITALY" and country_name != "UNITED KINGDOM": 
        plt.plot(All[country_name], linewidth = .7, alpha=.4, label=country_name)
    #plt.plot(All['Month'], MEX['count'], linewidth = 2.0, label="Mexico", color='#FFC60B')
plt.title("Search Frequencies of All Countries")
plt.xlabel("Months")
plt.ylabel("Number of Searches")
plt.tick_params(axis='x',bottom='off')
plt.tick_params(axis='y',left='off')
plt.show()
```
