```python
subset_100k = pd.read_csv("DF17/subset_data.csv")
```

```python
dest = pd.read_csv("DF17/dest.csv")
```

```python
dest_only = dest[['srch_destination_id','srch_destination_name']]
```

```python
subset_100k = subset_100k[["date_time", "hotel_country", "is_booking", "srch_destination_id"]]
subset_100k.dropna(inplace=True)
```

```python
subset_100k = subset_100k[subset_100k.is_booking != 0]
```

```python
subset_100k
```

```python
subset_100k = subset_100k.merge(dest_only,how='left',on='srch_destination_id')
```

```python
subset_100k['date_time'] = pd.to_datetime(subset_100k['date_time'])
subset_100k['date_time'] = subset_100k['date_time'].dt.date
subset_100k
```

```python
subset_100k['Month'] = pd.DatetimeIndex(subset_100k.date_time).month
subset_100k
```

```python
country_names = subset_100k.hotel_country.unique()
country_names
```
